```python
import pandas as pd
import numpy as np
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df_title = pd.read_csv("titles.csv")
df_credit = pd.read_csv("credits.csv")
```

```python
df_title.head()
```

```python
df_credit.head()
```

```python
df = pd.merge(df_title, df_credit, on="id")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df.genres.apply(pd.Series)
df.sample()
```

```python
df.drop("seasons", axis=1, inplace=True)
df.sample(5)
```

```python
df.isna().values.any()
```

```python
df.duplicated().values.any()
```

```python
data = df.dropna()
data.sample(5)
```

```python
data.info()
```

```python
type_cat = data.groupby("type", as_index=False).agg({"id":pd.Series.count})
type_cat.head()
```

```python
fig = px.pie(labels=type_cat.type, 
             values=type_cat.id,
             title="Movies vs Shows",
             names=type_cat.type,)
fig.update_traces(textposition='outside', textinfo='percent+label')
fig.show()
```

```python
genres_cat = data.groupby("genres", as_index=False).agg({"id":pd.Series.count})
genres_cat=genres_cat.sort_values("id", ascending=False)
genres_cat.head()
```

```python
bar = px.bar(x=genres_cat.genres, 
             y=genres_cat.id, 
             title='Top Genres',
             hover_name=genres_cat.genres,
             color=genres_cat.id,
             color_continuous_scale='Agsunset')

bar.update_layout(xaxis_title='Genre',
yaxis_title='Count',
coloraxis_showscale=False)

bar.show()
```

```python
rating = df.groupby("age_certification", as_index=False).agg({"id":pd.Series.count})
rating = rating.sort_values("id", ascending=False)
rating.head()
```

```python
rating_fig = px.pie(labels=rating.age_certification, 
             values=rating.id,
             title="Rating",
             names=rating.age_certification,)
rating_fig.update_traces(textposition='outside', textinfo='percent+label')
rating_fig.show()
```

```python
character = df.groupby("name", as_index=False).agg({"id":pd.Series.count})
character = character.sort_values("id", ascending=False)[:20]
character.head(10)
```

```python
char_bar = px.bar(x=character.name, 
             y=character.id, 
             title='Number of Movies/Shows performed by Actor',
             hover_name=character.name,
             color=character.id,
             color_continuous_scale='Agsunset')

char_bar.update_layout(xaxis_title='Name',
yaxis_title='Count',
coloraxis_showscale=False)

char_bar.show()
```

```python
country = df.groupby(["production_countries", "type"], as_index=False).agg({"id":pd.Series.count})
country = country.sort_values("id", ascending=False)[:50]
country.head(10)
```

```python
country_bar = px.bar(y=country.production_countries, 
             x=country.id,
             orientation="h",
             title='Top Countries ',
             hover_name=country.production_countries,
             color=country["type"],
             color_continuous_scale='Agsunset')

country_bar.update_layout(xaxis_title='Country',
yaxis_title='Count',
coloraxis_showscale=False)

country_bar.show()
```

```python
df_titles = df_title.dropna()
df_titles.head()
```

```python
top_rated = df_titles.sort_values("imdb_score", ascending=False)
top20_shows = top_rated[:20]
```

```python
top_bar = px.bar(x=top20_shows.title, 
             y=top20_shows.imdb_score, 
             title='Top 20 Netflix Shows',
             hover_name=top20_shows.title,
             color=top20_shows.imdb_score,
             color_continuous_scale='Agsunset')

top_bar.update_layout(xaxis_title='Name',
yaxis_title='Score',
coloraxis_showscale=False)

top_bar.show()
```
